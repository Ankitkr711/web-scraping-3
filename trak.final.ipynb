```python
import selenium
import pandas as pd 
from selenium import webdriver 
```

```python
from bs4 import BeautifulSoup
import requests
import time 
```

```python
driver = webdriver.Chrome('G:\\annaconda\\ankit\\chromedriver.exe')
```

```python
driver.get("https://trak.in/india-startup-funding-investment-2015/")
```

```python
company_date=[]
Company_name=[]
Company_location=[]
Company_deal=[]
```

```python
date=driver.find_elements_by_xpath("//td[@class='column-2']")
```

```python
for i in date:
    date=i.text
    company_date.append(date)
Date=company_date[34:64]
```

```python
Company=driver.find_elements_by_xpath("//td[@class='column-3']")

```

```python
for i in Company:
    Company=i.text
    Company_name.append(Company)
name=Company_name[34:64]
```

```python
Location=driver.find_elements_by_xpath("//td[@class='column-6']")
```

```python
for i in Location:
    Location=i.text
    Company_location.append(Location)
location=Company_location[34:64]
```

```python
Amount=driver.find_elements_by_xpath("//td[@class='column-9']")
```

```python
for i in Amount:
    Amount=i.text
    Company_deal.append(Amount)
amount=Company_deal[34:64]
```

```python
data=list(zip(Date,name,location,amount))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Date","Company","Location",'Deal'])
```
