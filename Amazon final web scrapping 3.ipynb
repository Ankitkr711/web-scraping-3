```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
Brand_name=[]
product_name=[]
Rating=[]
No_of_rating=[]
Price=[]
Return=[]
Expected_delivery=[]
availavilty=[]
Product_url=[]
for d in soup.findAll('div', attrs={'class':'a-section a-spacing-medium a-text-center'}):
        p=d.find('span', attrs={'class':'a-size-base-plus a-color-base'})
        Brand_name.append(p)
        pr=d.find('span', attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
        product_name.append(pr)
        ra=d.find('span', attrs={'class':'a-icon-alt'})
        Rating.append(ra)
        no=d.find('span', attrs={'class':'a-size-base'})
        No_of_rating.append(no)
        Pr=d.find('div', attrs={'class':'a-section a-spacing-none icon-content'})
        Price.append(Pr)
        Rt=d.find('a', attrs={'class':'a-link-normal a-text-normal'})
        Return.append(Rt)
        Ex=d.find('span', attrs={'class':'a-icon-alt'})
        Expected_delivery.append(Ex)
        
```
